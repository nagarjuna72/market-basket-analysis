```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import time
import sys
```

```python
data = pd.read_excel('DMT_DATA.xlsx', engine='openpyxl')

```

```python
#Remove spaces from the description column
data['Itemname'] = data['Itemname'].str.strip()

#Drop rows without invoice number
data.dropna(axis=0, subset=['BillNo'], inplace=True)
data['BillNo'] = data['BillNo'].astype('str')

#Remove the credit transaction with invoice numbers containing 'C'
data = data[~data['BillNo'].str.contains('C')]
data.head()
```

```python
basket = (data[data['Country'] =="France"]
          .groupby(['BillNo', 'Itemname'])['Quantity']
          .sum()
          .unstack()
          .reset_index()
          .fillna(0)
          .set_index('BillNo'))
basket.head(10)
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = basket.applymap(encode_units)
basket_sets.head()
```

APRIORI

```python
start_time_apriori = time.time()
frequent_itemsets_apriori = apriori(basket_sets, min_support=0.08, use_colnames=True)
end_time_apriori = time.time()
memory_usage_apriori = sys.getsizeof(frequent_itemsets_apriori)
num_itemsets_apriori = frequent_itemsets_apriori.shape[0]
print(frequent_itemsets_apriori)
```

```python
rules_apriori = association_rules(frequent_itemsets_apriori, metric="lift", min_threshold=1)
rules_apriori.head(10)
```

```python
rules_apriori[ (rules_apriori['lift'] >= 6) & (rules_apriori['confidence'] >= 0.8) ]
```

FP-GROWTH

```python
from mlxtend.frequent_patterns import fpgrowth
start_time_fpgrowth = time.time()
frequent_itemsets_fpgrowth = fpgrowth(basket_sets, min_support=0.07, use_colnames=True)
end_time_fpgrowth = time.time()
memory_usage_fpgrowth = sys.getsizeof(frequent_itemsets_fpgrowth)
num_itemsets_fpgrowth = frequent_itemsets_fpgrowth.shape[0]
print(frequent_itemsets_fpgrowth)
```

```python
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="lift", min_threshold=1)
rules_fpgrowth.head(10)
```

```python
rules_fpgrowth[ (rules_fpgrowth['lift'] >= 6) & (rules_fpgrowth['confidence'] >= 0.8) ]
```

ENSEMBLE

```python
start_time_ensemble = time.time()
frequent_itemsets_apriori = apriori(basket_sets, min_support=0.05, use_colnames=True)
frequent_itemsets_fpgrowth = fpgrowth(basket_sets, min_support=0.05, use_colnames=True)
ensemble_result = pd.concat([frequent_itemsets_apriori,frequent_itemsets_fpgrowth]).drop_duplicates().reset_index(drop=True)
end_time_ensemble = time.time()
memory_usage_ensemble = sys.getsizeof(ensemble_result)
num_itemsets_ensemble = ensemble_result.shape[0]
print(ensemble_result)
```

```python
rules_ensemble = association_rules(ensemble_result, metric="lift", min_threshold=1)
rules_ensemble.head(10)
```

```python
rules_ensemble[ (rules_ensemble['lift'] >= 6) & (rules_ensemble['confidence'] >= 0.8) ]
```

```python
print("Apriori:")
print(f"Number of frequent itemsets: {num_itemsets_apriori}")
print(f"Runtime: {end_time_apriori - start_time_apriori} seconds")
print(f"Memory usage: {memory_usage_apriori} bytes\n")

print("FP-Growth:")
print(f"Number of frequent itemsets: {num_itemsets_fpgrowth}")
print(f"Runtime: {end_time_fpgrowth - start_time_fpgrowth} seconds")
print(f"Memory usage: {memory_usage_fpgrowth} bytes\n")

print("Ensemble:")
print(f"Number of frequent itemsets: {num_itemsets_ensemble}")
print(f"Runtime: {end_time_ensemble - start_time_ensemble} seconds")
print(f"Memory usage: {memory_usage_ensemble} bytes")
```
